# Iris

At the Met Office we maintain a python library called Iris for working with gridded model data.

Iris implements a data model based on the [CF conventions](http://cfconventions.org/) giving you a powerful, format-agnostic interface for working with your data. It excels when working with multi-dimensional Earth Science data, where tabular representations become unwieldy and inefficient. 

Here is an example of loading and plotting some data in different projections using Iris.

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

import iris
import iris.plot as iplt
import iris.quickplot as qplt
```

```python
temperature = iris.load_cube('/s3/mogreps-g/prods_op_mogreps-g_20160917_06_19_153.nc', 'wet_bulb_potential_temperature')

# Plot #1: contourf with axes longitude from -180 to 180
plt.figure(figsize=(12, 5))
plt.subplot(121)
qplt.contourf(temperature[0], 15)
plt.gca().coastlines()

# Plot #2: contourf with axes longitude from 0 to 360
proj = ccrs.PlateCarree(central_longitude=-180.0)
plt.subplot(122, projection=proj)
qplt.contourf(temperature[0], 15)
plt.gca().coastlines()
iplt.show()
```
